```python
from random import randint
import math
import matplotlib.pyplot as plt

# Function to calculate Euclidean distance
def euclidean_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Assign points to clusters
def assign_to_clusters(data, cluster_centers):
    clusters = [[] for _ in range(len(cluster_centers))]
    for point in data:
        min_distance = float('inf')
        closest_cluster = None
        for i, center in enumerate(cluster_centers):
            distance = euclidean_distance(point, center)
            if distance < min_distance:
                min_distance = distance
                closest_cluster = i
        clusters[closest_cluster].append(point)
    return clusters

# Update cluster centers
def update_centers(clusters):
    new_centers = []
    for cluster in clusters:
        if cluster:
            center_x = sum(point[0] for point in cluster) / len(cluster)
            center_y = sum(point[1] for point in cluster) / len(cluster)
            new_centers.append((center_x, center_y))
        else:
            new_centers.append((float('inf'), float('inf')))  # Placeholder for empty clusters
    return new_centers

# Function to plot clusters
def plot_clusters(clusters, cluster_centers):
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
    for i, cluster in enumerate(clusters):
        color = colors[i % len(colors)]  # Wrap around colors if more clusters than colors
        plt.scatter(*zip(*cluster), color=color)
    plt.scatter(*zip(*cluster_centers), color='black', marker='x', label='Centroids')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('K-means Clustering')
    plt.legend()
    plt.show()

num_points = int(input("Enter number of data points: "))
x = [randint(0, 50) for _ in range(num_points)]
y = [randint(0, 50) for _ in range(num_points)]
data = list(zip(x, y))
print("Data points:", data)

num_clusters = int(input("Enter the number of clusters: "))

# Initialize cluster centers
cluster_centers = []

# Take input for cluster centers from the user
for i in range(num_clusters):
    print(f"Enter coordinates for cluster center {i+1}:")
    center_x = int(input("x-coordinate: "))
    center_y = int(input("y-coordinate: "))
    cluster_centers.append((center_x, center_y))

# K-means clustering
max_iterations = 100  # Maximum number of iterations
for _ in range(max_iterations):
    clusters = assign_to_clusters(data, cluster_centers)
    new_centers = update_centers(clusters)
    if new_centers == cluster_centers:  # Convergence check
        break
    cluster_centers = new_centers

# Print final clusters
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1}: {cluster}")

# Plot clusters
plot_clusters(clusters, cluster_centers)
```

```python
df=pd.read_csv('grocery.csv',names=["products"], header=None)

print(df)
```

```python
data = list(df["products"].apply(lambda x: x.split(',')))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
# Fitting the TransactionEncoder object to the data and transforming it
te_data=te.fit(data).transform(data)
# Creating a DataFrame from the transformed data with column names from TransactionEncoder
df=pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
# Applying the Apriori algorithm to the DataFrame df with specified parameters
  # Finds frequent itemsets with a minimum support of 2%
df1=apriori(df,min_support=0.02,use_colnames=True)
df1
```

```python
df1.sort_values(by="support",ascending=False)
```

```python
# Adding a new column "length" to the DataFrame df1
# The column "length" represents the number of items in each frequent itemset
df1["length"] = df1["itemsets"].apply(lambda x: len(x))
df1
```

```python
# Filtering the DataFrame df1 to select frequent itemsets with a length of 2 and support greater than 0.05
df1[(df1["length"]==3) & (df1["support"]>0.05)]#same can be done with 2/4
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming df1 contains the frequent itemsets obtained from the Apriori algorithm
# df1 should be in the format returned by the apriori function
rules = association_rules(df1, metric="confidence", min_threshold=0.5)
print(rules)
```

```python
rules = association_rules(df1, metric="confidence", min_threshold=0.8)

# Print the strong association rules
print("Strong Association Rules:")
print(rules)
```
