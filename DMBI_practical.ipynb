B. chi-square test

```python
#with inbuilt lib
import numpy as np
from scipy.stats import chi2_contingency

# Sample contingency table
observed = np.array([[1200, 600, 200],
                     [200, 500, 2300]])
# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Print results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
#check if attributes are related or not
alpha = 0.05
if p < alpha:
  print("the attributes are correlated")
else:
  print("the attributes are not correlated")
```

```python
#solving whole thing
import numpy as np
from scipy import stats

def calculate_expected(observed):
  #calculate expected values of each cell in contingency table
  row_totals = observed.sum(axis=1)
  col_totals = observed.sum(axis=0)
  total = observed.sum()

  expected = []
  for i in range(observed.shape[0]):
    expected_row =[]
    for j in range(observed.shape[1]):
      expected_value = (row_totals[i] * col_totals[j])/ total
      expected_row.append(expected_value)
    expected.append(expected_row)
  return np.array(expected)

def chi_square_test(observed):
  #calculate expected values
  expected = calculate_expected(observed)
  #calculate chi-square statistics
  chi_square_statistics = np.sum((observed - expected) ** 2 / expected)
  #calculate degrees of freedom
  df = (observed.shape[0] - 1) * (observed.shape[1] - 1)
  p_value= 5.991
  return chi_square_statistics, df, p_value
# create a contingency table (2x2 for simplicity)
observed_data = np.array([[1200, 600, 200], [200, 500, 2300]])

#perform chi-square test
chi_square_stat, degrees_of_freedom, p_value = chi_square_test(observed_data)

#display results
print(f"Chi-square statistics: {chi_square_stat}")
print(f"Degree of freedom: {degrees_of_freedom}")
print(f"P-value: {p_value}")

#check if correlated based on p-value
alpha = 0.05
if p_value < alpha:
  print("the attributes are correlated")
else:
  print("the attributes are not correlated")
```

```python
import numpy as np

def chi_square_test(observed):
    # Calculate expected values
    expected = np.outer(observed.sum(axis=1), observed.sum(axis=0)) / observed.sum()

    # Calculate chi-square statistics
    chi_square_statistics = np.sum((observed - expected) ** 2 / expected)

    # Calculate degrees of freedom
    df = (observed.shape[0] - 1) * (observed.shape[1] - 1)

    # P-value for alpha=0.05 and 1 degree of freedom
    p_value = 5.991

    return chi_square_statistics, df, p_value

# Create a contingency table
observed_data = np.array([[1200, 600, 200], [200, 500, 2300]])

# Perform chi-square test
chi_square_stat, degrees_of_freedom, p_value = chi_square_test(observed_data)

# Display results
print(f"Chi-square statistics: {chi_square_stat}")
print(f"Degree of freedom: {degrees_of_freedom}")
print(f"P-value: {p_value}")

# Check if correlated based on p-value
alpha = 0.05
if p_value < alpha:
    print("The attributes are correlated")
else:
    print("The attributes are not correlated")
```

```python
import sys
import matplotlib
matplotlib.use('Agg')

import pandas as pd
from sklearn import tree, datasets
import io
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("trees.csv")
data
```

```python
dIncome = {'Very high': 4, 'High' : 3, 'Medium': 2, 'Low' : 1}
data['Income'] = data['Income'].map(dIncome)
dAge = {'Old' : 3, 'Medium' : 2, 'Young' : 1}
data['Age'] = data['Age'].map(dAge)
dHouse = {'Yes' : 1, 'Rented' : 0}
data['Own house'] = data['Own house'].map(dHouse)
print(data)
```

```python
features = ['Income', 'Age', 'Own house']
X = data[features]
y = data['Own house']
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
tree.plot_tree(dtree, feature_names=features)
```

```python
#method one:
from sklearn.tree import plot_tree
from IPython.display import Image
# Plot and save the decision tree as an image
plt.figure(figsize=(10, 8))
plot_tree(dtree, feature_names=features, class_names=['No', 'Yes'], filled=True)
plt.savefig("decision_tree.png")
Image("decision_tree.png")
```

```python
#method 2:
from sklearn.tree import export_graphviz
import pydotplus
from sklearn.tree import plot_tree
from IPython.display import Image

# Export decision tree to DOT format
dot_data = export_graphviz(dtree, out_file=None,
                           filled=True, rounded=True,
                           special_characters=True, feature_names=features, class_names=['No', 'Yes'])

# Convert DOT data to PNG image
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png("decision_tree.png")

# Display the saved image
Image("decision_tree.png")
```
